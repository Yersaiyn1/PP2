```python
import pygame

pygame.init()

# настройки экрана 
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Extended Paint Application")

# цвет
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
ERASER_COLOR = WHITE

# Инструменты 
PENCIL = "pencil"  #ручка
RECTANGLE = "rectangle"  #Төртбұрыш
CIRCLE = "circle"  #круг
ERASER = "eraser"  #ластик

current_tool = PENCIL
drawing = False
start_pos = None
selected_color = BLACK
brush_size = 5

# Цвет фона 
screen.fill(WHITE)
pygame.display.flip()
#код для выбора цвета и формы с помощью клавиш
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:
                current_tool = PENCIL
            elif event.key == pygame.K_r:
                current_tool = RECTANGLE
            elif event.key == pygame.K_c:
                current_tool = CIRCLE
            elif event.key == pygame.K_e:
                current_tool = ERASER
            elif event.key == pygame.K_1:
                selected_color = RED
            elif event.key == pygame.K_2:
                selected_color = GREEN
            elif event.key == pygame.K_3:
                selected_color = BLUE 
            elif event.key == pygame.K_0:
                selected_color = BLACK

        elif event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            start_pos = event.pos
            if current_tool == PENCIL:
                pygame.draw.circle(screen, selected_color, event.pos, brush_size)

        elif event.type == pygame.MOUSEBUTTONUP:
            drawing = False
            if current_tool == RECTANGLE:
                end_pos = event.pos
                rect_width = end_pos[0] - start_pos[0]
                rect_height = end_pos[1] - start_pos[1]
                pygame.draw.rect(screen, selected_color, (start_pos[0], start_pos[1], rect_width, rect_height), 2)

            elif current_tool == CIRCLE:
                end_pos = event.pos
                radius = int(((end_pos[0] - start_pos[0]) ** 2 + (end_pos[1] - start_pos[1]) ** 2) ** 0.5)
                pygame.draw.circle(screen, selected_color, start_pos, radius, 2)

        elif event.type == pygame.MOUSEMOTION:
            if drawing:
                if current_tool == PENCIL:
                    pygame.draw.circle(screen, selected_color, event.pos, brush_size)
                elif current_tool == ERASER:
                    pygame.draw.circle(screen, ERASER_COLOR, event.pos, brush_size * 2)

    pygame.display.update()

pygame.quit()
```
