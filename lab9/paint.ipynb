```python
# Библиотеки 
import pygame
import sys
import math

pygame.init()

# Настройки экрана
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Extended Paint App")

# Цвета
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
COLORS = {
    pygame.K_1: BLACK,
    pygame.K_2: (255, 0, 0),      
    pygame.K_3: (0, 255, 0),       
    pygame.K_4: (0, 0, 255),       
    pygame.K_5: (255, 255, 0),      
    pygame.K_6: (255, 0, 255),     
    pygame.K_7: WHITE              
}

# Инструменты
PENCIL = "pencil"
RECTANGLE = "rectangle"
CIRCLE = "circle"
SQUARE = "square"
RIGHT_TRIANGLE = "right_triangle"
EQUILATERAL_TRIANGLE = "equilateral_triangle"
RHOMBUS = "rhombus"
ERASER = "eraser"

# Дефолтные параметры
current_color = BLACK
drawing = False
start_pos = None
current_tool = PENCIL
eraser_size = 10

 
screen.fill(WHITE)
clock = pygame.time.Clock()

def draw_shape(tool, start, end, color):
    x1, y1 = start
    x2, y2 = end

    if tool == RECTANGLE:
        pygame.draw.rect(screen, color, pygame.Rect(min(x1, x2), min(y1, y2), abs(x2 - x1), abs(y2 - y1)), 2)

    elif tool == CIRCLE:
        radius = int(math.hypot(x2 - x1, y2 - y1))
        pygame.draw.circle(screen, color, start, radius, 2)

    elif tool == SQUARE:
        size = min(abs(x2 - x1), abs(y2 - y1))
        pygame.draw.rect(screen, color, pygame.Rect(x1, y1, size, size), 2)

    elif tool == RIGHT_TRIANGLE:
        points = [start, (x2, y2), (x1, y2)]
        pygame.draw.polygon(screen, color, points, 2)

    elif tool == EQUILATERAL_TRIANGLE:
        side = abs(x2 - x1)
        height = int(side * math.sqrt(3) / 2)
        points = [start, (x1 + side, y1), (x1 + side / 2, y1 - height)]
        pygame.draw.polygon(screen, color, points, 2)

    elif tool == RHOMBUS:
        mid_x = (x1 + x2) // 2
        mid_y = (y1 + y2) // 2
        points = [(mid_x, y1), (x2, mid_y), (mid_x, y2), (x1, mid_y)]
        pygame.draw.polygon(screen, color, points, 2)

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        # Выбор цвета
        if event.type == pygame.KEYDOWN:
            if event.key in COLORS:
                current_color = COLORS[event.key]

            # Выбор инструментов с клавиш
            if event.key == pygame.K_p:
                current_tool = PENCIL
            elif event.key == pygame.K_r:
                current_tool = RECTANGLE
            elif event.key == pygame.K_c:
                current_tool = CIRCLE
            elif event.key == pygame.K_s:
                current_tool = SQUARE
            elif event.key == pygame.K_t:
                current_tool = RIGHT_TRIANGLE
            elif event.key == pygame.K_e:
                current_tool = EQUILATERAL_TRIANGLE
            elif event.key == pygame.K_h:
                current_tool = RHOMBUS
            elif event.key == pygame.K_x:
                current_tool = ERASER

         
        if event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            start_pos = event.pos

        elif event.type == pygame.MOUSEBUTTONUP:
            if drawing and current_tool != PENCIL and current_tool != ERASER:
                end_pos = event.pos
                draw_shape(current_tool, start_pos, end_pos, current_color)
            drawing = False

        elif event.type == pygame.MOUSEMOTION and drawing:
            if current_tool == PENCIL:
                pygame.draw.line(screen, current_color, start_pos, event.pos, 2)
                start_pos = event.pos
            elif current_tool == ERASER:
                pygame.draw.circle(screen, WHITE, event.pos, eraser_size)

    pygame.display.flip()
    clock.tick(60)
```
